```python
pat=r"(\-)"
tafe_resignations["institute_service"]=tafe_resignations["institute_service"].str.replace(pat," ").str.split().str[0].str.replace("Less","0").str.replace("More","20")
tafe_resignations["institute_service"]=tafe_resignations["institute_service"].astype(float)
#tafe_resignations["institute_service"]
#dete_resignations["institute_service"].astype(str)
#dete_resignations["institute_service"]
```

```python
def update_vals(val):
    if pd.isnull(val) == True:
        return np.nan
    elif val =="-":
        return False
    else:
        return True
```

```python
check_Job=tafe_resignations[["Contributing Factors. Dissatisfaction","Contributing Factors. Job Dissatisfaction"]].applymap(update_vals)
check_Job
```

```python
tafe_resignations["dissatisfied"]=check_Job.any(axis=1,skipna=False)
tafe_resignations
tafe_resignations_up=tafe_resignations
```

```python
check_dete=dete_resignations[["job_dissatisfaction","dissatisfaction_with_the_department","physical_work_environment","lack_of_recognition","lack_of_job_security","work_location","employment_conditions","work_life_balance","workload"]].applymap(update_vals)
check_dete
```

```python
dete_resignations["dissatisfied"]=check_dete.any(axis=1,skipna=False)
dete_resignations_up=dete_resignations
dete_resignations_up.columns
```

```python
dete_resignations_up["institue"]="DETE"
tafe_resignations_up["institute"]="TAFE"
combined=pd.concat([dete_resignations_up,tafe_resignations_up],axis=0)
combined.columns
```

```python
def func(val):
    if pd.isnull(val):
        return "No data"
    elif val <3:
        return "New"
    elif val>=3 and val<=6:
        return "Experienced"
    elif val >=7 and val <=10:
        return "Established"
    elif val>11:
        return "Veteran"
```

```python
combined["service_cat"]=combined["institute_service"].apply(func)
combined["service_cat"]
```

```python
combined["dissatisfied"].value_counts(dropna=False).sort_index()
```

```python
combined["dissatisfied"].fillna(True)
combined["dissatisfied"]=combined["dissatisfied"].apply(lambda val:1 if(val) else 0)
```

```python
final=combined.pivot_table(index="service_cat",values="dissatisfied")
```

```python
%matplotlib inline
final.plot(kind="bar")
```
